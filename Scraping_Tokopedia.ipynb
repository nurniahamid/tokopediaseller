```python
from google_play_scraper import reviews, Sort

scrape_review, _ = reviews(
    'com.tokopedia.sellerapp',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=11000
)
```

```python
import csv

with open('review_tokopedia_seller.csv', mode='w', newline='', encoding='utf-8') as file :
  writer = csv.writer(file)
  writer.writerow(['Review'])
  for review in scrape_review :
    writer.writerow([review['content']])
```
